```python
import pandas as pd
import seaborn as sns 
import numpy as np
```

```python
df=pd.read_csv("C:/3-statistics/income(1).csv",index_col=None,names=["income","count"],skiprows=1)
df.head()
```

```python
df
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)}) #rc stands for "runtime configuration," and 'figure.figsize':(11.7,8.27) 
                                            #specifies the dimensions of the figure in inches (width x height)
g = sns.barplot(x='income',y='count',data=df)  #data=df: Specifies that the data for the plot comes from the DataFrame df.
g.set_xticklabels(g.get_xticklabels(),rotation=60,horizontalalignment='right'); 
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
g = sns.barplot(x='income',y='count',data=df)
g.set_xticklabels(g.get_xticklabels(),rotation=45,horizontalalignment='right');
g.set(xscale='log')
```
